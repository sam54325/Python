```python
lst = [2, 3, 'Py', '10', 1, 'SQL', 5.5, True, 3, 'John', None, 7]
datatypes = {'int': 0, 'str': 0, 'float': 0, 'bool': 0, 'NoneType': 0}
for i in lst:
    datatypes[type(i).__name__] += 1
print(datatypes)
```

```python
li = [5, 7, 22, 97, 54, 62, 77, 23, 73, 61]
odd_numbers = []
for num in li:
    if num % 2 != 0:
        odd_numbers.append(num)
print(odd_numbers)
```

```python
def area(shape, **kwargs):
    if shape == 'square':
        if 'side' in kwargs:
            return kwargs['side'] ** 2
        else:
            return "Please enter 'side' for a square"
    elif shape == 'rectangle':
        if 'length' in kwargs and 'width' in kwargs:
            return kwargs['length'] * kwargs['width']
        else:
            return "Please enter 'length' and 'width' for a rectangle"
    elif shape == 'circle':
        if 'radius' in kwargs:
            return 3.1416 * kwargs['radius'] ** 2
        else:
            return "Please enter 'radius' for a circle"
    else:
        return "Invalid shape"

print(area('square', side=4))
print(area('rectangle', length=5, width=3))
print(area('circle', radius=7))
```

```python
def is_prime(num):
    if num < 2:
        return False
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            return False
    return True

primes = [str(num) for num in range(2, 101) if is_prime(num)]
print(','.join(primes))
```
